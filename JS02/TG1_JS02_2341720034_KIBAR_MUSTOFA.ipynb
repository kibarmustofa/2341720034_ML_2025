## Memisahkan variabel

Pisahkan fitur (variabel yang dapat digunakan) dari target (variabel yang tidak dapat digunakan untuk prediksi, dalam kasus ini 'diagnosis').


```python
# Separate features (X) and target (y)
df = df.drop(['id', 'Unnamed: 32'], axis=1)
y = df['diagnosis']

display(df.head())
display(y.head())
```

## Encoding Target

Lakukan encoding pada kolom 'diagnosis' menggunakan `LabelEncoder`.

```python
# Encode the target variable
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head(10)
```

## Standardisasi Fitur

Lakukan standardisasi pada semua kolom numerik dalam fitur menggunakan `StandardScaler`.

```python
std = StandardScaler()
# Select numerical columns from X
numerical_cols = df.columns
df[numerical_cols] = std.fit_transform(df[numerical_cols])
display(df.head(10))
```
